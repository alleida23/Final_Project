```python
def book_recommender(book_searched, gr_data):
    """
    Recommends books similar to the one searched for by the user.
    """
    # Remove common words from the search term
    common_words = ['the', 'and', 'a', 'an', 'in', 'of', 'to', 'at']
    book_searched = ' '.join([word for word in book_searched.lower().split() if word not in common_words])

    # Check if search term is at least 3 characters long
    if len(book_searched) < 3:
        message = "Please enter a longer search term."
        display(Markdown(message))
        return None

    # Find books with matching title or author
    title_matches = gr_data[gr_data['title'].str.lower().str.contains(book_searched)]
    author_matches = gr_data[gr_data['author'].str.lower().str.strip() == book_searched]

    # Combine title and author matches and calculate match score
    matches = pd.concat([title_matches, author_matches]).drop_duplicates()
    matches['match_score'] = matches['title'].apply(lambda x: fuzz.partial_ratio(book_searched, x.lower()))

    # Sort matches by match score in descending order
    matches = matches.sort_values(by='match_score', ascending=False)

    # Display top 10-11 books with highest match score
    if matches.empty:
        message = f"Sorry, we couldn't find any books similar to **{book_searched.capitalize()}**. Please try again."
        display(Markdown(message))
    else:
        message = f"Here are the top {min(11, len(matches))} books similar to **{book_searched.capitalize()}**:\n"
        for i, match in enumerate(matches.head(11)['title']):
            message += f"{i+1}. {match}\n"
        display(Markdown(message))

    return None
```
