```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import random
```

```python
# URL of the Pulitzer Prize winners for the Novel category
url = "https://www.pulitzer.org/prize-winners-by-category/261"

# Send a GET request to the URL and store the response object
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

response.status_code
```

```python
# Find all the year links
year_links = soup.find_all("a", class_="ng-binding ng-scope")
```

```python
# Extract the year from each link
years = []
for link in year_links:
    year = link.text.strip()
    years.append(year)

print(years)
```
